# Libraries

```python
import pandas as pd
import json
```

# Tabular Phase

```python
prob_dist = 'uniform' #['poisson','exponential','uniform_int','uniform']

f = open(f'data/raw_data/{prob_dist}.json')

data = json.load(f)
data
```

```python
df1_batch1 = pd.DataFrame([data[0]])
df1_batch2 = pd.DataFrame([data[1]])

df2_batch1 = df1_batch1.explode('values').reset_index(drop=True)
df2_batch2 = df1_batch2.explode('values').reset_index(drop=True)
```

```python
df2_batch1
```

```python
df2_batch2
```

```python
# salva em .csv
df2_batch1.to_csv(f'./data/tabular_data/{prob_dist}_batch1.csv', sep = ',', decimal='.', header=True)
df2_batch2.to_csv(f'./data/tabular_data/{prob_dist}_batch2.csv', sep = ',', decimal='.', header=True)
```
